```python
from astropy.io import ascii
from astropy.coordinates import SkyCoord
from astropy.coordinates import FK5
from astropy import units as u
import numpy as np
import pandas as pd

cat3 = ascii.read("CAT3.1")

print(np.shape(cat3))

cat3_ra = cat3['RAJ2000']
cat3_dec = cat3['DEJ2000']

cat  = SkyCoord(ra=cat3_ra*u.degree, dec=cat3_dec*u.degree, frame='fk5', equinox='J2000.0')
cat3_J2016 = cat.transform_to(FK5(equinox='J2016.0'))

print(cat3_J2016)
```

```python
sky_coords = []
for i in range(len(cat3_J2016)):
  sky_coords.append([cat3_J2016[i].ra.degree,cat3_J2016[i].dec.degree])
print(sky_coords)

RAJ2016=[x[0] for x in sky_coords]
DECJ2016=[x[1] for x in sky_coords]

print(RAJ2016)
print(DECJ2016)
```

```python
cat3csv = pd.read_csv('CAT3.1')
cat3csv.insert(loc=1, column='RAJ2016', value=RAJ2016, allow_duplicates=True)
cat3csv.insert(loc=2, column='DECJ2016', value=DECJ2016, allow_duplicates=True)
cat3csv.to_csv("Cat31precessed.csv")
```
